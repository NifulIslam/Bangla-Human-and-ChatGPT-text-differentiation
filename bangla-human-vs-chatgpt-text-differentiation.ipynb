```python
import numpy as np
import pandas as pd
df = pd.read_csv("/kaggle/input/chatgpt-dataset/merge dataset V2.csv")
df
```

```python
df = df.drop_duplicates()
df
```

```python
melted_df = pd.melt(df, 
                    id_vars=['Category'], 
                    value_vars=['Context(Human)', 'Context(ChatGpt 1st paraphrase)'], 
                    var_name='source', 
                    value_name='text')

# Drop rows where text is NaN
melted_df = melted_df.dropna(subset=['text'])

# Create 'label' column based on 'source'
melted_df['label'] = melted_df['source'].apply(lambda x: 'human' if 'Human' in x else 'chatgpt')

# Drop the 'source' column as it's no longer needed
melted_df = melted_df.drop(columns=['source'])

# Reset index for neatness
melted_df = melted_df.reset_index(drop=True)
df = melted_df
df = df.drop('Category', axis=1)
df = df.sample(frac=1)
df
```

```python
def count_tokens(text):
    return len(text.split())

df['token_count'] = df['text'].apply(count_tokens)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.kdeplot(df['token_count'], shade=True)
plt.xlabel('Token Count')
plt.ylabel('Density')
plt.show()
```

```python
import statistics
mean = sum(df['token_count']) / len(df['token_count']) #mean
std = statistics.pstdev(df['token_count']) #Standard deviation
mean,std
```

```python
max_len= int(mean+ (3*std))
max_len
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Bidirectional, LSTM, Dense, GlobalMaxPooling1D, AveragePooling1D, Flatten, MaxPooling1D, Dropout, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['label'])

# Split dataset
texts = df['text'].values
labels = df['label'].values
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Tokenize and pad sequences
max_words = 20000

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train)
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)
```

```python
input_layer = Input(shape=(max_len,))
embedding_layer = Embedding(max_words, 128)(input_layer)

lstm_layer_1 = Bidirectional(LSTM(64, return_sequences=True))(embedding_layer)

lstm_layer_2 = Bidirectional(LSTM(32, return_sequences=True))(lstm_layer_1)

lstm_layer_3 = Bidirectional(LSTM(16, return_sequences=True))(lstm_layer_2)

gap = GlobalAveragePooling1D()(lstm_layer_3)
dense_layer_1 = Dense(8, activation='relu')(gap)

dense_layer_2 = Dense(4, activation='relu')(dense_layer_1)
output_layer = Dense(1, activation='sigmoid')(dense_layer_2)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
from sklearn.metrics import matthews_corrcoef, f1_score, precision_score, recall_score
import matplotlib.pyplot as plt

history = model.fit(X_train_pad, y_train, epochs=10, batch_size=512, validation_split=0.1)


y_pred_lstm = model.predict(X_test_pad)
y_pred_lstm = (y_pred_lstm > 0.5).astype(int)

lstm_accuracy = accuracy_score(y_test, y_pred_lstm)
lstm_mcc = matthews_corrcoef(y_test, y_pred_lstm)
lstm_f1 = f1_score(y_test, y_pred_lstm)
lstm_precision = precision_score(y_test, y_pred_lstm)
lstm_recall = recall_score(y_test, y_pred_lstm)

print(f'Bi-LSTM Model Accuracy: {lstm_accuracy}')
print(f'Bi-LSTM Model MCC: {lstm_mcc}')
print(f'Bi-LSTM Model F1 Score: {lstm_f1}')
print(f'Bi-LSTM Model Precision: {lstm_precision}')
print(f'Bi-LSTM Model Recall: {lstm_recall}')


plt.figure(figsize=(12, 4))


plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')


plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.show()
```

```python
feature_extractor = Model(inputs=model.input, outputs=gap) 
train_features = feature_extractor.predict(X_train_pad)
test_features = feature_extractor.predict(X_test_pad)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

param_grid = {
    'kernel': ['rbf', 'linear','sigmoid'],
    'gamma': [0.1, 0.5, 1, 1.2, 2],
    'C': [0.1, 1, 10, 100,200,500]
}


svm = SVC()


grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)


grid_search.fit(train_features, y_train)


best_svm = grid_search.best_estimator_
```

```python
pr = best_svm.predict(test_features)


acc_score = metrics.accuracy_score(y_test, pr)
pre_score = metrics.precision_score(y_test, pr, average="weighted")
recall = metrics.recall_score(y_test, pr, average="weighted")
f1 = metrics.f1_score(y_test, pr, average="weighted")
mcc = metrics.matthews_corrcoef(y_test, pr)

print(f'Best Parameters: {grid_search.best_params_}')
print(f'Accuracy: {acc_score}')
print(f'Precision: {pre_score}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(f'Matthews Correlation Coefficient: {mcc}')

# Plot confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, pr)
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Greens', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python

print("Best Parameters found: ", grid_search.best_params_)
print("Best Cross-validation score: ", grid_search.best_score_)


results = grid_search.cv_results_

print("\nGrid Search Results:")
for mean, std, params in zip(results['mean_test_score'], results['std_test_score'], results['params']):
    print(f"Mean Accuracy: {mean:.3f} Std Dev: (+/- {std * 2:.3f}) for {params}")



results_df = pd.DataFrame(results)

print("\nDetailed Grid Search Results (first 10 rows):")
print(results_df.head(10))
```
